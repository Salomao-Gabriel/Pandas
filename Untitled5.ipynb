```python
import csv

# Define a dictionary mapping US state names to numbers
us_state_numbers = {
    'Alabama': '2',
    'Alaska': '3',
    'Arizona': '4',
    'Arkansas': '5',
    'California': '6',
    'Colorado': '7',
    'Connecticut': '8',
    'Delaware': '9',
    'Florida': '10',
    'Georgia': '11',
    'Hawaii': '12',
    'Idaho': '13',
    'Illinois': '14',
    'Indiana': '15',
    'Iowa': '16',
    'Kansas': '17',
    'Kentucky': '18',
    'Louisiana': '19',
    'Maine': '20',
    'Maryland': '21',
    'Massachusetts': '22',
    'Michigan': '23',
    'Minnesota': '24',
    'Mississippi': '25',
    'Missouri': '26',
    'Montana': '27',
    'Nebraska': '28',
    'Nevada': '29',
    'New Hampshire': '30',
    'New Jersey': '31',
    'New Mexico': '32',
    'New York': '33',
    'North Carolina': '34',
    'North Dakota': '35',
    'Ohio': '36',
    'Oklahoma': '37',
    'Oregon': '38',
    'Pennsylvania': '39',
    'Rhode Island': '40',
    'South Carolina': '41',
    'South Dakota': '42',
    'Tennessee': '43',
    'Texas': '44',
    'Utah': '45',
    'Vermont': '46',
    'Virginia': '47',
    'Washington': '48',
    'West Virginia': '49',
    'Wisconsin': '50',
    'Wyoming': '51'
}

# Read the CSV file
input_file = 'input.csv'
output_file = 'output.csv'

with open(input_file, 'r') as file:
    reader = csv.DictReader(file)
    rows = list(reader)

# Modify the "region" column
for row in rows:
    state = row['region']
    if state in us_state_numbers:
        row['region'] = us_state_numbers[state]

# Replace empty values in the "Position" column with "Watching"
for row in rows:
    if not row['Position']:
        row['Position'] = 'Watching'

# Write the modified data to a new CSV file
fieldnames = reader.fieldnames
with open(output_file, 'w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(rows)

print("CSV file processed successfully!")
```
